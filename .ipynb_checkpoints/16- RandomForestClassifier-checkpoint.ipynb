```python
X = df.drop('income',axis=1)
y = df['income']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X , y, test_size = 0.3, random_state= 0)
```

```python
categorical = [col for col in X_train.columns if X_train[col].dtype == "O" ] # object ise 
```

```python
categorical
```

```python
X_train[categorical].isnull().sum()
```

```python
# fill the nan datas
X_train['workclass'].mode()[0] # can fill with this data
```

```python
for i in [X_train, X_test]:
    i['workclass'] = i['workclass'].fillna(X_train['workclass'].mode()[0])
    i['occupation'] = i['occupation'].fillna(X_train['occupation'].mode()[0])
    i['native-country'] = i['native-country'].fillna(X_train['native-country'].mode()[0])
```

```python
X_train[categorical].isnull().sum()
```

```python
X_test[categorical].isnull().sum()
```

```python
# encoding 
```

```python
X_train[categorical].head()
```

```python
df[categorical].nunique()
```

```python
df[['native-country','income']]
```

```python
y_train_binary = y_train.apply(lambda x: 1 if x.strip() == '>50K' else 0)
```

```python
target_means = y_train_binary.groupby(X_train['native-country']).mean()
```

```python
X_train['native-country-encoded'] = X_train['native-country'].map(target_means)
X_train['native-country-encoded'] = X_train['native-country-encoded'].fillna(y_train_binary.mean())


X_test['native-country-encoded'] = X_test['native-country'].map(target_means)
X_test['native-country-encoded'] = X_test['native-country-encoded'].fillna(y_train_binary.mean())
```

```python
X_train.head()
```

```python
# now we can drop native-country column
X_train.drop('native-country',axis=1,inplace=True)
#X_test.drop('native-country',axis=1,inplace=True)
```

```python
X_test.head()
```

```python
categorical
```

```python
# onehot encoding for the other columns
one_hot_categories = [
    'workclass',
     'education',
     'marital-status',
     'occupation',
     'relationship',
     'race',
     'sex']
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#  sparse_output=False -> not create matrix create np array
encoder = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown="ignore",sparse_output=False), one_hot_categories)
    ],
    remainder="passthrough"
)
```

```python
X_train.head()
```

```python
X_train_encoded = encoder.fit_transform(X_train)
X_test_encoded = encoder.transform(X_test)
```

```python
X_train_encoded
```

```python
# we can get columns from encoder
columns = encoder.get_feature_names_out()
```

```python
#columns
```

```python
# we can convert our encoded arrays to df again
X_train = pd.DataFrame(X_train_encoded, columns = columns, index = X_train.index)
X_test = pd.DataFrame(X_test_encoded, columns = columns, index = X_test.index)
```

```python
#X_train
```

```python
#X_train.describe()
```

```python
# X_train.info() all float
```

```python
cols = X_train.columns
```

```python
# RobustScaler - resist to outliers
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# training
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=10, random_state=15)
rfc.fit(X_train_scaled,y_train)
```

```python
y_pred = rfc.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
# let's try with 100 
rfc = RandomForestClassifier(n_estimators=100, random_state=15)
rfc.fit(X_train_scaled,y_train)
y_pred2 = rfc.predict(X_test_scaled)
print(accuracy_score(y_test,y_pred2)) # 0.85!
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```
