```python
X = df.drop('Species',axis=1)
y = df['Species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=15)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train_scaled,y_train)
```

```python
y_pred = gnb.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print("confusion matrix: \n",confusion_matrix(y_test,y_pred))
print("accuracy_score: \n",accuracy_score(y_test,y_pred))
print("classification_report: \n",classification_report(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
# t-SNE
from sklearn.manifold import TSNE

X_embedded = TSNE(n_components=2, random_state=42).fit_transform(X_test_scaled)
plt.scatter(X_embedded[:,0], X_embedded[:,1], c=y_test, cmap='coolwarm', edgecolor='k', s=60)
plt.title('t-SNE projection of classes')
plt.show()
```

```python
#PCA ile 2D veya t-SNE ile 2D/3D projeksiyon
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_test_2d = pca.fit_transform(X_test_scaled)

plt.figure(figsize=(8,6))
plt.scatter(X_test_2d[:,0], X_test_2d[:,1], c=y_test, cmap='coolwarm', edgecolor='k', s=60)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Projection of Classes')
plt.show()
```

```python
#hyperparameter tuning with class weights to handle imbalance
penalty=['l1', 'l2', 'elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
params=dict(penalty=penalty,C=c_values,solver=solver)

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
grid=GridSearchCV(estimator=LogisticRegression(multi_class='ovr'),param_grid=params,scoring='accuracy',cv=cv)

import warnings
warnings.filterwarnings('ignore')

grid.fit(X_train_scaled,y_train)
```

```python
y_pred = grid.predict(X_test_scaled)
```

```python
y_pred
```

```python
print("score: ",accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
confusion_matrix(y_pred,y_test) 
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
from sklearn.preprocessing import label_binarize
y_test_bin = label_binarize(y_test, classes=np.unique(y_test))
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np

y_prob = grid.predict_proba(X_test_scaled)  # shape = (n_samples, n_classes)
classes = np.unique(y_test)

plt.figure(figsize=(8,6))

for i, cls in enumerate(classes):
    fpr, tpr, _ = roc_curve(y_test_bin[:,i], y_prob[:,i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'Class {cls} (AUC = {roc_auc:.2f})')

plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multiclass ROC Curve (OvR)')
plt.legend()
plt.show()
```
